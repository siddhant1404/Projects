```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("https://api.covid19india.org/csv/latest/raw_data1.csv")
df2=pd.read_csv("https://api.covid19india.org/csv/latest/raw_data2.csv")
df3=pd.read_csv("https://api.covid19india.org/csv/latest/raw_data3.csv")
df4=pd.read_csv("https://api.covid19india.org/csv/latest/raw_data4.csv")
df1=df1.append(df2)
df1=df1.append(df3)
df1=df1.append(df4)
df1.reset_index(inplace=True)
df1=df1.drop(["index"],axis=1)
```

```python
dfworld=pd.read_csv("owid-covid-data.csv")
dfworld=dfworld[["location","date","total_cases","new_cases","total_tests","new_tests"]]
df1=dfworld[dfworld["location"]=="India"]
df1=df1.fillna(0)
df1.set_index("location",inplace=True)
df=pd.DataFrame()
df=df1
df.iloc[148,3]=df.iloc[147,3] + df.iloc[147,4]
df.iloc[149,3]=df.iloc[148,3] 
df.iloc[150,3]=df.iloc[148,3] 
df=df.append(df.iloc[150,:])
df
```

```python
fig,ax=plt.subplots()
observation_dates = np.arange('2019-12-31','2020-05-31', dtype='datetime64[D]')
observation_dates1 = np.arange('2019-12-31','2020-05-31', dtype='datetime64[D]')
plt.plot(observation_dates,(df7[["total_cases"]]))
plt.plot(observation_dates,(df7[["total_tests"]]))
plt.plot(observation_dates1,df[["total_cases","total_tests"]])
plt.legend(['RUSSIA total cases', 'RUSSIA total tests','INDIA total cases','INDIA total tests'])
plt.gca().fill_between((observation_dates), 
                       df7["total_cases"],df7["total_tests"],
                       facecolor="red", 
                       alpha=0.5)
plt.gca().fill_between((observation_dates), 
                       df["total_cases"],df["total_tests"],
                       facecolor="blue", 
                       alpha=0.5)
plt.ylabel("CASES IN MILLIONS")
plt.xlabel("MONTHS")
plt.axhline(y=173763,label="India",color="blue")
plt.axhline(y=387623,label="Russia",color="red")
plt.axhline(y=9671394,label="Russia",color='red')
plt.axhline(y=3358201,label="India",color="blue")
fig.tight_layout()
plt.show()
```
